```python
sampling_rate = [10,20,30,60,120]
fixadr = 'map-matching-dataset'
numbers = os.listdir('map-matching-dataset')

#w개수를 다르게
fmm_config = fmm_config = {'k':50,'r':0.002,'gps_error':0.0001,'beta':0.0001,'w':8}

result_my = [[0]*len(sampling_rate) for i in range(len(numbers))]
result_org = [[0]*len(sampling_rate) for i in range(len(numbers))]


for i,number in enumerate(numbers):
    print(number)
    read_adr = fixadr+"\\" + number + "\\" + number +'.'
    #model_adr : pickle로 저장한 model을 위한 주소. rtree이슈로 사용안함.
    model_adr = fixadr + '\\' + number +"\\" +'model.p'
    org, my = get_rmf(read_adr,model_adr, sampling_rate, fmm_config)
    
    for j in range(len(sampling_rate)):
        result_my[i][j] = my[j]
    for j in range(len(sampling_rate)):
        result_org[i][j] = org[j]
    
with open('result_org.pickle','wb') as file:
    pickle.dump(result_org,file)
with open('result_my.pickle','wb') as file:
    pickle.dump(result_my,file)
```
