```python
from model import *
from rdp_time_experiment import * # 전처리 적용했을 때의 시간 체크
#from non_rdp_time_experiment import * # 전처리 전후를 비교하기 위해서 rdp적용 x.
import os
import pickle
```

```python
sampling_rate = [10,20,30,60,120]
fixadr = 'map-matching-dataset'
numbers = os.listdir('map-matching-dataset')

#w개수를 다르게
fmm_config = fmm_config = {'k':50,'r':0.002,'gps_error':0.0001,'beta':0.0001,'w':6}

result_my = [[0]*len(sampling_rate) for i in range(len(numbers))]
result_org = [[0]*len(sampling_rate) for i in range(len(numbers))]

total_time_my = 0.0
total_time_org = 0.0

for i,number in enumerate(numbers):
    print(number)
    read_adr = fixadr+"\\" + number + "\\" + number +'.'
    #model_adr : pickle로 저장한 model을 위한 주소. rtree이슈로 사용안함.
    model_adr = fixadr + '\\' + number +"\\" +'model.p'
    org, my, org_time, my_time = get_rmf(read_adr,model_adr, sampling_rate, fmm_config)
    total_time_my += my_time
    total_time_org += org_time
    
    for j in range(len(sampling_rate)):
        result_my[i][j] = my[j]
    for j in range(len(sampling_rate)):
        result_org[i][j] = org[j]
    
with open('result_org.pickle','wb') as file:
    pickle.dump(result_org,file)
with open('result_my.pickle','wb') as file:
    pickle.dump(result_my,file)
with open('result_org_time.pickle','wb') as file:
    pickle.dump(total_time_org,file)
with open('result_my_time.pickle','wb') as file:
    pickle.dump(total_time_my,file)
```
